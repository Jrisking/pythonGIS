```python
from scipy.integrate import simpson
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
# 数据来自于matlab doc
vel = np.array([0, 0.45, 1.79, 4.02, 7.15, 11.18, 16.09, 21.90, 29.05, 29.05,
29.05, 29.05, 29.05, 22.42, 17.9, 17.9, 17.9, 17.9, 14.34, 11.01,
8.9, 6.54, 2.03, 0.55, 0])
time = np.arange(vel.shape[0])
```

```python
fig, ax = plt.subplots(figsize=(8, 4), dpi=120)
ax.plot(time, vel, '-*')
ax.set_title("Automobile Velocity")
ax.set_xlabel("Time (s)")
ax.set_ylabel("Velocity (m/s)")
```

### 计算总行驶距离

```python
fig, ax = plt.subplots(figsize=(8, 4), dpi=120)
ax.plot(time, vel, '-*')
ax.set_title("Automobile Velocity")
ax.set_xlabel("Time (s)")
ax.set_ylabel("Velocity (m/s)")
ax.fill_between(time, vel, 0, alpha=0.3)
```

```python
simpson(vel, time)
```
