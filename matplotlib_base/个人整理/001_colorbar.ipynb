## 设置colorbar刻度的标签
1. 如何添加一个colorbar
2. 如何设置colorbar的刻度
3. 如何设置colorbar刻度的标签（字体大小、旋转角度等）

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm
from numpy.random import randn


# Fixing random state for reproducibility
np.random.seed(19680801)
```

```python
%matplotlib widget
```

```python
fig, ax = plt.subplots()

data = np.clip(randn(250, 250), -1, 1)

cax = ax.imshow(data, cmap=cm.coolwarm)
ax.set_title('Gaussian noise with vertical colorbar')

# Add colorbar, make sure to specify tick locations to match desired ticklabels
cbar = fig.colorbar(cax, ticks=[-1, 0, 1])
cbar.ax.set_yticklabels(['< -1', '0', '> 1'])  # vertically oriented colorbar
```

```python
fig, ax = plt.subplots(constrained_layout=True)

data = np.clip(randn(250, 250), -1, 1)

cax = ax.imshow(data, cmap=cm.afmhot)
ax.set_title('Gaussian noise with horizontal colorbar')

# shrink 参数: 让colorbar变小一点
cbar = fig.colorbar(cax, ticks=[-1, 0, 1], orientation='horizontal', shrink=0.6)

cbar.ax.set_xticklabels(['Low', 'Medium', 'High'],
                        rotation=40, # 让标签旋转
                        fontdict={'size': 20} # 设置标签的大小等
                        )  # horizontal colorbar

plt.show()
```

##### 离散型标签

```python
fig, ax = plt.subplots(constrained_layout=True)

data = np.random.randint(0, 4, size=(42, 100))
cmap = cm.get_cmap('Blues', np.unique(data).shape[0])
cax = ax.imshow(data, cmap=cmap, vmin=data.min()-0.5, vmax=data.max()+0.5)
ax.set_title('discrete colorbar')

cbar = fig.colorbar(cax, ticks=[0, 1, 2, 3], shrink=0.4)
cbar.ax.set_yticklabels([f'class{i}' for i in range(4)]) 

# Add colorbar, make sure to specify tick locations to match desired ticklabels
# cbar = fig.colorbar(cax, ticks=[-1, 0, 1])
# cbar.ax.set_yticklabels(['< -1', '0', '> 1'])  # vertically oriented colorbar
```
